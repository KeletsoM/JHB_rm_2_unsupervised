### Proportionality of User Ratings

To view proportions of ratings categories, it is best practice to use pie charts.
Below it can be seen that the 4.0 - 5.0 rating carries the most weight at almost 50%. This is an indication that most people who rate movies, are likely to give a higher rating. 

```python
# Calculate and categorise ratings proportions into a human-readable format
a = len(train.loc[train['rating']== 0.5]) / len(train)
b = len(train.loc[train['rating']==1.0]) / len(train)
c = len(train.loc[train['rating']==1.5]) / len(train)
d = len(train.loc[train['rating']==2.0]) / len(train)
low_ratings= a+b+c+d
e = len(train.loc[train['rating']==2.5]) / len(train)
f = len(train.loc[train['rating']== 3.0]) / len(train)
g = len(train.loc[train['rating']==3.5]) / len(train)
medium_ratings= e+f+g
h = len(train.loc[train['rating']==4.0]) / len(train)
i = len(train.loc[train['rating']==4.5]) / len(train)
j = len(train.loc[train['rating']==5.0]) / len(train)
high_ratings= h+i+j
# Sanity check for checking if ratings proportions equal 100 percent
low_ratings+medium_ratings+high_ratings==1
```

```python
# Creating labels 
labels = 'Low Ratings (scale: 0.5 - 2.0)', 'Medium Ratings (scale: 2.5 - 3.5)', 'High Ratings (scale: 4.0 - 5.0)'
sizes = [low_ratings, medium_ratings,  high_ratings]
explode = (0, 0, 0.1)  # Only "explore" the 3rd slice (i.e. 'Anti')

# Create pie chart with above labels and calculated class proportions as inputs 
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=270)

# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal') 
plt.title('Categorised Proportions of User Ratings ',bbox={'facecolor':'k',
                                                           'pad':5},color='w'
                                                           ,fontsize = 18)
plt.show()
```

```python
# Ranking top movies based on viewership
def popularity(train,title,ids):
    train_data_grouped = train.groupby([title])[ids].count().reset_index()  #user_id  #movie title
    train_data_grouped.rename(columns = {ids: 'Viewers'},inplace=True)            
    train_data_sort = train_data_grouped.sort_values(['Viewers',title], ascending = [0,1])
    train_data_sort['Rank'] = train_data_sort['Viewers'].rank(ascending=0, method='first')
    popularity_recommendations = train_data_sort.head(10) 
    return popularity_recommendations
```

### The long tail problem

Below it can be seen that the probability that a user only rates very few movies is high. The movies that are most likely to be rated are probably the more popular movies. At this point we can see that there is a 'long tail' to the right of the probability density function (PDF) and the cumulative probability density function (CDF). This is known as the 'long-tail' problem where we have a disproportionately large number of movies that have not been rated yet relative to movies that have been rated. Some assumptive reasons for this could be that popular movies became popular through word-of-mouth advertising and other forms movie-budget-based advertising. Nonetheless there could be some really great movies out there that are similar to the type of movies a user would be curious to see and to ultimately rate. But because of the high search costs(i.e. time spent looking for a movie), a user will most probably never get to know about the latent movie that sits along the long tail of the PDF or CDF below. Of course this is a scenario without the use of a recommender system. With the use of a recommender system, the user will be made aware of a universe of similarly preferred movies to what they have watched and then have a chance to rate them. Theerefore the below curve substantiates the need for a recommender system and confirms that we have a long tail problem our movie data. 

Imagine being an investor in a movie production and after having invested so much money, you never get ratings on whether your movie performed well or not because people did not actually know that it existed for a long period of time. What are the chances that you will make informed investment decisions going into the the future for similar projects? Probably next to none. However, ratings data could form part of your key performance indicator (KPI) metrics for determining whether a particular movie investment project is likely to be successful based on certain movie features to help you make investment decisions more confidently.

```python
# Method to make axes more readable
def human(num, units = 'M'):
    units = units.lower()
    num = float(num)
    if units == 'k':
        return str(num/10**3) + 'K'
    elif units == 'm':
        return str(num/10**6) + 'M'
    elif units == 'b':
        return str(num/10**9) +  'B'
```
