# Introduction

The purpose of this report is to contstruct a recommender system which will accurately predict how a user will rate a movie they have not yet seen based on historical preferences. The economic value in producing such a system has great commercial impact because it will mean users will enjoy the service which eliminates the process of having to search for your ideal movie to watch, creating brand loyality and also increased revenues as more movies being watched.

Because this is unsupervised learning, there is no target response in the data set and thus there are many predictors which will have to be used to identify the relevance of fit to the recommender algorithm. These include genre, rating, budget, director and release date to name a few. 

For this report, an investigation has to be conducted as to whether a content-based or collaborative filtering approach will be used and therefore, it will be prudent to test both methods and use the one which produces the most accurate results.

## Sparsity matrix creation for training set

```python
# Create disk-based sparsity matrix for training data to preserve memory for certain aspects of notebook
if os.path.isfile('train_sparse_matrix.npz'):
    print("It is present in your pwd, getting it from disk....")
    # just get it from the disk instead of computing it
    train_sparse_matrix = sparse.load_npz('train_sparse_matrix.npz')
    print("DONE..")
else: 
    print("We are creating sparse_matrix from the dataframe..")
    # create sparse_matrix and store it for after usage.
    # csr_matrix(data_values, (row_index, col_index), shape_of_matrix)
    # It should be in such a way that, MATRIX[row, col] = data
    train_sparse_matrix = sparse.csr_matrix((train.rating.values, (train.userId.values,
                                               train.movieId.values)),)
    
    print('Done. It\'s shape is : (user, movie) : ',train_sparse_matrix.shape)
    print('Saving it into disk for further usage..')
    # save it into disk
    sparse.save_npz("train_sparse_matrix.npz", train_sparse_matrix)
    print('Done..\n')
```

```python
# Create user(us), movie (mv) columns and determine NULL count for training data

us,mv = train_sparse_matrix.shape
elem = train_sparse_matrix.count_nonzero()

print("Sparsity Of Train matrix: {} % ".format(  (1-(elem/(us*mv))) * 100) )
```

## EDA

After perusing our data we reflect back on our problem statement in our [introduction](#Introduction) where since this is an unsupervised learning problem, there is no target response in the data set and thus there are many predictors which will have to be used to identify the relevance of fit to the recommender algorithm. These include genre, rating, budget, director and release date to name a few. We can explore some assumptions. For example, we can expect only a small subset of movies to have a disproportionately high number of ratings compared to the majority of movies and more popular movies will most likely get rated the most. We will explore our data to verify this assumption and also explore additional trends that will help us to form some more assumptions along the way. We expect to have computational peformance (algorithmic time complexity) challenges as we work with large data sets and we expect to work on not only our local machines but also on cloud-based servers that have enough computational power to ingest our data sets that we expect to be processor and memory hungry. To read more about time complexity, click [here](https://towardsdatascience.com/understanding-time-complexity-with-python-examples-2bda6e8158a7)
.  Alternatively we can transform our data sets into a compressed format where can still get the key features of our data but keep it in a manageable size for data processing purposes. Once we  have our data in a manageable size and we have our EDA, we can form insights to confirm our assumptions and we can start asking questions about which model will be best suited to predicting user movie ratings given the aforementioned challenges and what we know about the trends in the data at that point.


Since we are using training data that is historical in nature, we begin analysis by delving into some time-based aspects of  the MovieLens rating data.
```python
# Use sparsity matrix toet the user averages in dictionary (key: user_id/movie_id, value: avg rating)

def get_average_ratings(sparse_matrix, of_users):
    
    # average ratings of user/axes
    ax = 1 if of_users else 0 # 1 - User axes,0 - Movie axes

    # ".A1" is for converting Column_Matrix to 1-D numpy array 
    sum_of_ratings = sparse_matrix.sum(axis=ax).A1
    # Boolean matrix of ratings ( whether a user rated that movie or not)
    is_rated = sparse_matrix!=0
    # no of ratings that each user OR movie..
    no_of_ratings = is_rated.sum(axis=ax).A1
    
    # max_user  and max_movie ids in sparse matrix 
    u,m = sparse_matrix.shape
    # creae a dictonary of users and their average ratigns..
    average_ratings = { i : sum_of_ratings[i]/no_of_ratings[i]
                                 for i in range(u if of_users else m) 
                                    if no_of_ratings[i] !=0}

    # return that dictionary of average ratings
    return average_ratings
```

### The long tail problem

Below it can be seen that the probability that a user only rates very few movies is high. The movies that are most likely to be rated are probably the more popular movies. At this point we can see that there is a 'long tail' to the right of the probability density function (PDF) and the cumulative probability density function (CDF) as a continuation of the blue plot for the user average ratings and movie average ratings mentioned further above. This is known as the 'long-tail' problem where we have a disproportionately large number of movies that have not been rated yet relative to movies that have been rated. Some assumptive reasons for this could be that popular movies became popular through word-of-mouth advertising and other forms movie-budget-based advertising. Nonetheless there could be some really great movies out there that are similar to the type of movies a user would be curious to see and to ultimately rate. But because of the high search costs(i.e. time spent looking for a movie), a user will most probably never get to know about the latent movie that sits along the long tail of the PDF or CDF below. Of course this is a scenario without the use of a recommender system. With the use of a recommender system, the user will be made aware of a universe of similarly preferred movies to what they have watched and then have a chance to rate them. Theerefore the below curve substantiates the need for a recommender system and confirms that we have a long tail problem our movie data. 

Imagine being an investor in a movie production and after having invested so much money, you never get ratings on whether your movie performed well or not because people did not actually know that it existed for a long period of time. What are the chances that you will make informed investment decisions going into the the future for similar projects? Probably next to none. However, ratings data could form part of your key performance indicator (KPI) metrics for determining whether a particular movie investment project is likely to be successful based on certain movie features to help you make investment decisions more confidently.

Below, we take a subset of 100 users from our sparse matrix to give us an indication of how may hidden movie ratings or latent factors can exist amongst 100 users and the movies they have rated. This is a difficult computational challenge that may require a lot of compute power, so we have to start thinking of an algorithm that is best suited to solve this challenge optimally i.e. getting an accurate recommendation and prediction without being too slow in getting these results for a user.

```python
# Create sparse matrix using a subset of 100 users from the training set

from sklearn.metrics.pairwise import cosine_similarity


def compute_user_similarity(sparse_matrix, compute_for_few=False, top = 100, verbose=False, verb_for_n_rows = 20,
                            draw_time_taken=True):
    no_of_users, _ = sparse_matrix.shape
    # get the indices of  non zero rows(users) from our sparse matrix
    row_ind, col_ind = sparse_matrix.nonzero()
    row_ind = sorted(set(row_ind)) # we don't have to
    time_taken = list() #  time taken for finding similar users for an user..
    
    # we create rows, cols, and data lists.., which can be used to create sparse matrices
    rows, cols, data = list(), list(), list()
    if verbose: print("Computing top",top,"similarities for each user..")
    
    start = datetime.now()
    temp = 0
    
    for row in row_ind[:top] if compute_for_few else row_ind:
        temp = temp+1
        prev = datetime.now()
        
        # get the similarity row for this user with all other users
        sim = cosine_similarity(sparse_matrix.getrow(row), sparse_matrix).ravel()
        # We will get only the top ''top'' most similar users and ignore rest of them..
        top_sim_ind = sim.argsort()[-top:]
        top_sim_val = sim[top_sim_ind]
        
        # add them to our rows, cols and data
        rows.extend([row]*top)
        cols.extend(top_sim_ind)
        data.extend(top_sim_val)
        time_taken.append(datetime.now().timestamp() - prev.timestamp())
        if verbose:
            if temp%verb_for_n_rows == 0:
                print("computing done for {} users [  time elapsed : {}  ]"
                      .format(temp, datetime.now()-start))
            
        
    # lets create sparse matrix out of these and return it
    if verbose: print('Creating Sparse matrix from the computed similarities')
    #return rows, cols, data
    
    if draw_time_taken:
        plt.title("Sparse matrix creation duration",bbox={'facecolor':'k', 'pad':5},color='w',fontsize = 16)
        plt.plot(time_taken, label = 'time taken for each user')
        plt.plot(np.cumsum(time_taken), label='Total time')
        plt.legend(loc='best')
        plt.xlabel('User')
        plt.ylabel('Time (seconds)')
        plt.show()
        
    return sparse.csr_matrix((data, (rows, cols)), shape=(no_of_users, no_of_users)), time_taken
from datetime import datetime

d1 = datetime.now()
start = d1
u_u_sim_sparse, _ = compute_user_similarity(train_sparse_matrix, compute_for_few=True, top = 100,
                                                     verbose=True)
print("-"*100)

print("Time taken :",datetime.now()-start)
```

```python
# Use tried and tested multifactorization algorithm (SVD) to explain latent factors that exist for 100 users

from datetime import datetime
from sklearn.decomposition import TruncatedSVD

start = datetime.now()

trunc_svd = TruncatedSVD(n_components=700, algorithm='randomized', random_state=15)
trunc_svd = trunc_svd.fit(train_sparse_matrix)

print(datetime.now()-start)
```

## Alternating Least Squares (ALS) algorithm

Even though we expect FunK SVD, we attempt to start our modelling by using ALS for the scalability reasons mentioned above. ALS is a method used in collaborative filtering used in recommendation systems. ALS is a matrix factorization technique its called alternating because it uses the concept of latent factors, in a recommender system there are multiple of characteristics that can tie people and products together, the role of the recommender is to take advantage of these latent factors to recommend specific products to people.For a more clearer picture it is noticeable that the phrase "Least Squares" is rather familiar from linear regression, in the case of alternating the "least squares" phrase is due to the fact that during one iteration one latent factor is kept constant while the other changes then in the next iteration the other latent factor is kept constant and the other is varied, while at the same time minimising the sum of the squared errors. ALS ties together the concept of linear regression and grouping from K-means clustering.
